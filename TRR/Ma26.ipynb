```python
def permutations_with_replacement (a, n):
    r = len(a)
    if sum (n) == 0:
        return [[]]
    P = []
    for i in range(r):
        if n [i] > 0:
            n_ = n.copy()
            n_[i] -= 1
            for p in permutations_with_replacement (a, n_):
                p = [a[i]] + p
                P.append(p)
    return P

permutations_with_replacement (['A', 'B', 'L'], [3, 2, 3])
```
